```python
from PyQt5 import QtWidgets
from PyQt5.QtWidgets import QFileDialog
import sys
'''
pyqt5(anaconda自带)，可以完成界面操作
'''
from openpyxl import Workbook
from openpyxl import load_workbook
'''
openpyxl可以操纵excel文件
'''
import os
from win32com.client import Dispatch
import win32com.client
import time
import win32api,win32con
'''
win32com可以执行window命令
win32api也可以执行window命令
'''

class MyWindow(QtWidgets.QWidget):
    def __init__(self):
        super(MyWindow,self).__init__()
        self.myButton = QtWidgets.QPushButton(self)
        self.myButton.setObjectName("myButton")
        self.myButton.setText("选取文件")
        self.myButton.clicked.connect(self.msg)
        '''
        编辑gui对话框
        '''
    def msg(self):
        directory1 = QFileDialog.getExistingDirectory(self,"选取文件夹","./")
        newline = directory1[::-1]
        position=newline.find('/')
        loo = newline[0:position]
        school = loo[::-1]
        fileName2 = QFileDialog.getOpenFileName(self,"选取放置结果的文件","./")
        '''
        选择放置结果的excel名称和相应的根目录
        如’C:/Users/Administrator/values.xlsx‘
        注意是反斜杠,"./"代表用/分割
        '''
        rootDir = directory1
        value_file = fileName2[0]  #原因同上
        startnum =1
        for filename in os.listdir(rootDir):
            location_1 = filename.index('.')
            newfilename = int(filename[:location_1])
            '''
            通过os包遍历文件夹中所有的文件名
            '''
            excel_name = rootDir+"/"+filename
            '''
            生成根目录+文件名的形式，便于后面使用
            '''
            a = []
            a.append(startnum)
            a.append(school)
            wb = load_workbook(excel_name,data_only=True)
            sheet = wb.get_sheet_names()
            ws = wb[sheet[9]]
            a.append(ws.cell(row = 2,column = 6).value)
            ws1 = wb[sheet[10]]
            a.append(ws1.cell(row = 2,column = 8).value)
            ws2 = wb[sheet[11]]
            a.append(ws2.cell(row = 2,column = 9).value)
            ws3 = wb[sheet[12]]
            a.append(ws3.cell(row = 2,column = 6).value)
            ws4 = wb[sheet[13]]
            a.append(ws4.cell(row = 13,column = 1).value)
            ws5 = wb[sheet[14]]
            a.append(ws5.cell(row = 2,column = 8).value)
            ws5 = wb[sheet[15]]
            a.append(ws5.cell(row = 3,column = 16).value)
            ws6 = wb[sheet[16]]
            a.append(ws6.cell(row = 2,column = 10).value)
            ws7 = wb[sheet[17]]
            a.append(ws7.cell(row = 3,column = 13).value)

            wb3 = load_workbook(value_file,data_only = True)
            sheet = wb3.get_sheet_names()
            ws3 = wb3.worksheets[0]
            rownum = ws3.max_row+1
            startnum = startnum +1
            for j in range(1,12):
                ws3.cell(row = rownum,column = j).value = a[j-1]
            wb3.save(value_file)

if __name__=="__main__": 
    import sys 
    app=QtWidgets.QApplication(sys.argv) 
    myshow=MyWindow()
    myshow.show()
    sys.exit(app.exec_()) 
```
